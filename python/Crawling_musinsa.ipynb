```python
brands, names, prices, stars, nums, images = [],[],[],[],[],[]
time.sleep(3)

for i in range(1,51):
    brand = driver.find_elements_by_css_selector("#searchList > li:nth-child("+str(i)+") > div.li_inner > div.article_info > p.item_title > a")
    brands.append(brand[0].text)
    name = driver.find_elements_by_css_selector("#searchList > li:nth-child("+str(i)+") > div.li_inner > div.article_info > p.list_info > a")
    names.append(name[0].text)
    price = driver.find_elements_by_css_selector("#searchList > li:nth-child("+str(i)+") > div.li_inner > div.article_info > p.price")
    prices.append(int(re.sub("[,]", "", price[0].text.split()[-1][:-1])))    # price[0].text.split()[-1][:-1].replace(',','')
    star = driver.find_elements_by_css_selector("#searchList > li:nth-child("+str(i)+") > div.li_inner > div.article_info > p.point > span.img-score")
    if star == []:
        stars.append(0)
    else:
        stars.append(int(star[0].get_attribute("class")[-2:]))
#     star = driver.find_elements_by_xpath("#searchList > li:nth-child("+str(i)+") > div.li_inner > div.article_info > p.point > span.img-score")
    num = driver.find_elements_by_css_selector("#searchList > li:nth-child("+str(i)+") > div.li_inner > div.article_info > p.point > span.count")
    if num == []:
        nums.append(0)
    else:
        nums.append(num[0].text)
    image = driver.find_elements_by_css_selector("#searchList > li:nth-child("+str(i)+") > div.li_inner > div.list_img > a > img")
    images.append(image[0].get_attribute("src"))
```

```python
musinsa = pd.DataFrame([brands, names, prices, stars, nums, images]).T
musinsa.columns=['brands', 'names', 'prices', 'stars', 'nums', 'images']
```
